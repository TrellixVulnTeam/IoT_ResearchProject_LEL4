## Collapse into minutes
* Use in-built Pandas function
* `resample('min').max()` 

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
ds = pd.read_csv('S1SubActivities_timeRangeBoolean.csv', index_col = 'duration')
ds.index = pd.to_datetime(ds.index)
```

```python
ds.head(n=2)
```

```python
ds.tail(n=2)
```

```python
ds = ds.resample('min').max()
```

```python
ds.head(n=2)
```

```python
ds.tail(n=2)
```

```python
len(ds)
```

```python
ds.to_csv('S1SubActivities_timeRangeBooleanMinutes.csv', index = 'duration')
```

```python
ds = ds.dropna(how='all')
```

```python
len(ds)
```

```python
ds.to_csv('S1SubActivities_timeRangeBooleanMinutesDropNA.csv', index = 'duration')
```
